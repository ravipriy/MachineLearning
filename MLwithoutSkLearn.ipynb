```python
import csv
import math

class Node:
    def __init__(self, feature=None, threshold=None, result=None, left=None, right=None):
        self.feature = feature
        self.threshold = threshold
        self.result = result
        self.left = left
        self.right = right

# Load and preprocess the dataset from 'salaries.csv'
def load_dataset(filename):
    dataset = []
    with open(filename, 'r') as file:
        csv_reader = csv.reader(file)
        next(csv_reader)  # Skip the header row
        for row in csv_reader:
            company, job, degree, salary = row
            salary = int(salary == '1')  # Convert 'salary_more_than_100k' to binary (0 or 1)
            dataset.append([company, job, degree, salary])
    return dataset

# Split the dataset based on a feature and threshold
def split_dataset(dataset, feature, threshold):
    left, right = [], []
    for data_point in dataset:
        if data_point[feature] < threshold:
            left.append(data_point)
        else:
            right.append(data_point)
    return left, right

# Calculate Gini impurity for a dataset
def gini_impurity(dataset):
    total_samples = len(dataset)
    if total_samples == 0:
        return 0

    positive_samples = sum(data_point[-1] for data_point in dataset)
    p1 = (positive_samples / total_samples) ** 2
    p2 = (1 - positive_samples / total_samples) ** 2

    return 1 - (p1 + p2)

# Calculate the information gain for a split
def information_gain(left, right, current_uncertainty):
    p = len(left) / (len(left) + len(right))
    return current_uncertainty - p * gini_impurity(left) - (1 - p) * gini_impurity(right)

# Build the decision tree recursively
def build_tree(dataset, max_depth, depth=0):
    if depth >= max_depth:
        return Node(result=most_common_class(dataset))

    if len(set(data_point[-1] for data_point in dataset)) == 1:
        return Node(result=dataset[0][-1])

    best_feature, best_threshold, best_score, left, right = find_best_split(dataset)
    
    if best_score == 0:
        return Node(result=most_common_class(dataset))
    
    left_subtree = build_tree(left, max_depth, depth + 1)
    right_subtree = build_tree(right, max_depth, depth + 1)

    return Node(feature=best_feature, threshold=best_threshold, result=most_common_class(dataset), left=left_subtree, right=right_subtree)

# Find the best split for the dataset
def find_best_split(dataset):
    best_feature, best_threshold, best_score, left, right = None, None, 0, None, None
    current_uncertainty = gini_impurity(dataset)
    num_features = len(dataset[0]) - 1

    for feature in range(num_features):
        values = set(data_point[feature] for data_point in dataset)
        for value in values:
            left_subset, right_subset = split_dataset(dataset, feature, value)
            if len(left_subset) == 0 or len(right_subset) == 0:
                continue
            gain = information_gain(left_subset, right_subset, current_uncertainty)
            if gain > best_score:
                best_feature, best_threshold, best_score, left, right = feature, value, gain, left_subset, right_subset

    return best_feature, best_threshold, best_score, left, right

# Find the most common class in a dataset
def most_common_class(dataset):
    return max(set(data_point[-1] for data_point in dataset), key=list(data_point[-1] for data_point in dataset).count)

# Make a prediction using the decision tree
def predict(tree, data_point):
    if tree.result is not None:
        return tree.result
    if data_point[tree.feature] < tree.threshold:
        return predict(tree.left, data_point)
    else:
        return predict(tree.right, data_point)

# Load the dataset
dataset = load_dataset('salaries.csv')

# Build the decision tree
tree = build_tree(dataset, max_depth=3)

# Example prediction for a new data point (company='google', job='sales executive', degree='bachelors')
new_data_point = ['google', 'sales executive', 'bachelors']
prediction = predict(tree, new_data_point)

if prediction == 1:
    print("Predicted salary is more than $100K.")
else:
    print("Predicted salary is not more than $100K.")
```
